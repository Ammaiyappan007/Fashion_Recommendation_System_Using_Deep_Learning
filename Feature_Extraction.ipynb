```python
import tensorflow 
from keras.applications.resnet import ResNet50, preprocess_input
from keras.layers import GlobalMaxPooling2D
```

```python
model = ResNet50(weights="imagenet", include_top=False, input_shape=(224,224,3))
model.trainable = False
```

```python
model.summary()
```

```python
model = tensorflow.keras.Sequential([
    model,
    GlobalMaxPooling2D()
])
model.summary()
```

```python
def extract_featrure(img_path, model):
    img = cv2.imread(img_path)
    #img = cv2.resize(img,(224,224),fx=0,fy=0, interpolation = cv2.INTER_CUBIC)
    img = cv2.resize(img, (224,224))
    img_shape = np.array(img)
    expanded_img = np.expand_dims(img_shape, axis=0)
    pre_img = preprocess_input(expanded_img)
    result = model.predict(pre_img).flatten()
    normalized = result/norm(result)
    return normalized
```

```python
extract_featrure("D:\\Dataset\\5426.jpg", model)
```

```python
Extracted = []
feature_list = []

for file in os.listdir("D:\\Dataset"):
    Extracted.append(os.path.join("D:\\Dataset",file))
```

```python
Extracted
```

```python
for file in tqdm(Extracted):
    feature_list.append(extract_featrure(file, model))
```

```python
import pickle

pickle.dump(feature_list, open("featurevector.pkl", "wb"))
pickle.dump(Extracted,open("filenames.pkl", "wb"))
```
